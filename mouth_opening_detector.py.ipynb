```python
import cv2
import dlib
from scipy.spatial import distance


def is_mouth_open(shape):
    # nh ngha cc landmark lin quan n ming
    mouth_landmarks = list(range(48, 61))
    rate = shape.part(63).x - shape.part(67).x
#     print("rate", rate)
    
    # Tnh ton khong cch dc gia trung bnh ca cc landmark trn v di ming
    upper_mouth = sum(shape.part(i).y for i in range(48, 54)) / 6
    lower_mouth = sum(shape.part(i).y for i in range(54, 60)) / 6
    mouth_height = lower_mouth - upper_mouth
    if rate>0:
        RATE = mouth_height/rate
    else:
        RATE = 1.6
    print(rate)
    
    
    # iu kin kim tra ming c m hay khng
    return RATE > 1.5


font = cv2.FONT_HERSHEY_SIMPLEX 
COUNTER = 0
counter = 0
EYE_AR_CONSEC_FRAMES = 48



cap = cv2.VideoCapture(0)
hog_face_detector = dlib.get_frontal_face_detector()
dlib_facelandmark = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")


while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = hog_face_detector(gray)
    
    for face in faces:

        face_landmarks = dlib_facelandmark(gray, face)
        
        
#         draw_marks(img, shape[48:])
        
        cv2.putText(frame, 'Press r to record Mouth distances', (30, 30), font,
                    1, (0, 255, 255), 2)
        cv2.imshow("Output", frame)
    if cv2.waitKey(1) & 0xFF == ord('r'):
            
        break
cv2.destroyAllWindows()



while(True):
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = hog_face_detector(gray)
    
    if len(faces) == 0:
        cv2.putText(frame,"Please show your face!",(20,100),
            cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),4)
        
        counter += 1
        print(counter)
        if counter >= EYE_AR_CONSEC_FRAMES:
            cv2.putText(frame,"Unfocused!",(20,400),
                cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),4)
        print("unfocused")
    else: 
        counter = 0
    
    for face in faces:
        
        face_landmarks = dlib_facelandmark(gray, face)
        cnt_outer = 0
        cnt_inner = 0
        outnerMouth = []
        innerMouth = []
        for n in range(48, 60):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            outnerMouth.append((x,y))
            next_point = n+1
            if n == 59:
                next_point = 48
            x2 = face_landmarks.part(next_point).x
            y2 = face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x2,y2),(0,255,0),1)
            
        for n in range(60, 68):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            innerMouth.append((x,y))
            next_point = n+1
            if n == 67:
                next_point = 60
            x2 = face_landmarks.part(next_point).x
            y2 = face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x2,y2),(0,255,0),1)
            
        # Kim tra xem ming c m hay khng
        mouth_open = is_mouth_open(face_landmarks)

        # V cc landmark ln nh  kim tra
#         for i in range(68):
#             cv2.circle(frame, (face_landmarks.part(i).x, face_landmarks.part(i).y), 1, (0, 0, 255), -1)

        # Hin th kt qu
        if mouth_open:
            cv2.putText(frame, "Mouth open", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        else:
            cv2.putText(frame, "Mouth closed", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow("Output", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
